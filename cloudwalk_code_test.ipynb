# 2 - Consulta SQL e Gráfico
Criação de uma consulta em SQL e crie um gráfico dela e tente explicar o comportamento da anomalia que você encontrou.

```python
# Simulando uma consulta SQL com pandas, pois não temos um banco de dados relacional disponível.
# Suponha que 'checkout_1' e 'checkout_2' estejam em um banco de dados e queiramos comparar as vendas de 'hoje' em ambos.

# Renomeando a coluna 'Today' para evitar conflito na junção
checkout_1.rename(columns={'Today': 'Sales_Today_1'}, inplace=True)
checkout_2.rename(columns={'Today': 'Sales_Today_2'}, inplace=True)

# Criando um DataFrame com as vendas de 'hoje' de ambos os arquivos
# Use 'time' as the common column for the merge, as 'Hour' does not exist
sales_today = pd.merge(checkout_1[['time', 'yesterday']], checkout_2[['time', 'yesterday']], on='time', how='outer')

# Preenchendo valores nulos com 0, assumindo que não houve vendas naquela hora
sales_today.fillna(0, inplace=True)

# Criando um gráfico de linhas para comparar as vendas de 'hoje'
plt.figure(figsize=(10, 6))
plt.plot(sales_today['time'], sales_today['yesterday_x'], label='Ontem') # Use 'yesterday_x' from checkout_1
plt.plot(sales_today['time'], sales_today['yesterday_y'], label='Hoje') # Use 'yesterday_y' from checkout_2
plt.xlabel('Hora')
plt.ylabel('Vendas')
plt.title('Comparação de Vendas de "Hoje" em checkout_1 (Ontem) e checkout_2 (Hoje)')
plt.legend()
plt.grid(True)
plt.show()
```
